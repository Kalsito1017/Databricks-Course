```python
df = spark.read.csv("/Workspace/Users/Databricks-Course/customer.csv", inferSchema=True, header=True)
display(df)
```

```python
# Inner join the Premium customers DataFrame with the full DataFrame on 'customer_id'
df1 = df.filter(df.customer_type == "Premium")
df_inner = df1.join(df, on='customer_id', how='inner')
display(df_inner)
```

```python
# Left join the Premium customers DataFrame with the full DataFrame on 'customer_id'
df_left = df1.join(df, on='customer_id', how='left').filter(df1.customer_type == "Premium")
display(df_left)

# Right join the Premium customers DataFrame with the full DataFrame on 'customer_id'
df_right = df1.join(df, on='customer_id', how='right').filter(df1.customer_type == "Premium")
display(df_right)
```

```python
from pyspark.sql.functions import broadcast

df3 = df1.join(broadcast(df), on='customer_id', how='inner')
df3.explain()
display(df3)
```
