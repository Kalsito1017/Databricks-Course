```python
df = spark.read.csv("/Workspace/Users/Databricks-Course/customer.csv", inferSchema=True, header=True)
display(df)
```

```python
#spark.sql("create database retaildb")
#spark.sql("show databases")
#display(spark.sql("show databases"))##
df.write.mode("overwrite").format("delta").saveAsTable("customer_delta")
display(spark.sql("describe extended customer_delta"))
```

```python
spark.sql("""
    INSERT INTO retaildb.customer_delta VALUES 
    (1001, 'TimeSlayer', 'mygmail@gmail.com', 'NY', 'Premium', '2025-11-22', '23', 'Male', 100, current_timestamp())
""")
display(spark.sql("select * from retaildb.customer_delta"))
```

```python
spark.sql("update retaildb.customer_delta set customer_type = 'FAME' where customer_id = 1001")
display(spark.sql("select * from retaildb.customer_delta"))
```

```python
spark.sql("DELETE FROM retaildb.customer_delta WHERE customer_id = 1001")
display(spark.sql("SELECT * FROM retaildb.customer_delta"))
```
